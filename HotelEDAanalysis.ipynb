```python
import pandas as pd

# Convert 'arrival_date' to datetime format
df['arrival_date'] = pd.to_datetime(df['arrival_date'])

# Filter DataFrames by customer type
customerTransient = df[df['customer_type'] == 'Transient']
customerContract = df[df['customer_type'] == 'Contract']
customerTransientParty = df[df['customer_type'] == 'Transient-Party']
customerGroup = df[df['customer_type'] == 'Group']

# Set 'arrival_date' as index
customerTransient = customerTransient.set_index("arrival_date")
customerContract = customerContract.set_index("arrival_date")
customerTransientParty = customerTransientParty.set_index("arrival_date")
customerGroup = customerGroup.set_index("arrival_date")

# Ensure only numerical columns are summed
def resample_and_sum(df):
    numerical_cols = df.select_dtypes(include=['number']).columns
    return df[numerical_cols].resample('M').sum()

customerTransientMonthly = resample_and_sum(customerTransient)
customerContractMonthly = resample_and_sum(customerContract)
customerTransientPartyMonthly = resample_and_sum(customerTransientParty)
customerGroupMonthly = resample_and_sum(customerGroup)

fig = go.Figure()

# Plotting Transient Guests
if 'total_guest' in customerTransientMonthly.columns:
    fig.add_trace(go.Scatter(
        x=customerTransientMonthly.index,
        y=customerTransientMonthly['total_guest'],
        name="Transient Guests"
    ))

# Plotting Contract Guests
if 'total_guest' in customerContractMonthly.columns:
    fig.add_trace(go.Scatter(
        x=customerContractMonthly.index,
        y=customerContractMonthly['total_guest'],
        name="Contract Guests"
    ))

# Plotting Transient-Party Guests
if 'total_guest' in customerTransientPartyMonthly.columns:
    fig.add_trace(go.Scatter(
        x=customerTransientPartyMonthly.index,
        y=customerTransientPartyMonthly['total_guest'],
        name="Transient-Party Guests"
    ))

# Plotting Group Guests
if 'total_guest' in customerGroupMonthly.columns:
    fig.add_trace(go.Scatter(
        x=customerGroupMonthly.index,
        y=customerGroupMonthly['total_guest'],
        name="Group Guests"
    ))

fig.update_layout(
    title_text='Number of Guests by Customer Type',
    title_x=0.5,
    title_font=dict(size=30),
    xaxis_title="Arrival Date",
    yaxis_title="Number of Guests"
)

img_bytes = fig.to_image(format='png')

# Display the image
display(Image(img_bytes))
```

```python
from statsmodels.tsa.arima.model import ARIMA
resort_weekly_values = no_of_bookings_week_resort.values

# Perform ADF test
result_city = adfuller(resort_weekly_values)
print(result_city)

# Splitting the data into training and test sets
trainResort = resort_weekly_values[:90]
testResort = resort_weekly_values[90:]

# Define and fit the ARIMA model
modelResort = ARIMA(trainResort,  # Directly use the array
                    order=(2, 0, 0),
                    seasonal_order=(2, 1, 0, 12))

resultResort = modelResort.fit()

# Predicting the Test data
predictionsResortTest = resultResort.predict(start=90, end=113, typ='levels')

# Convert predictions to a Pandas Series and rename
predictionsResortTest = pd.Series(predictionsResortTest, name="Predictions")

print(predictionsResortTest)
```
